```python
import glob
import pandas as pd
import numpy as np
import ast
```

```python
annotations=glob.glob('data/OPP-115/annotations/*')
```

```python
annotations_list=[]
for file in annotations:
    annotations_list.append(pd.read_csv(file, header=None))
```

```python
annotations_df=pd.concat(annotations_list, ignore_index=True)
```

```python
annotations_df.head()
annotations_df.shape
```

```python
annotations_df.columns=['annotation_ID', 'batch_ID', 'annotator_ID', 'policy_ID', 'segment_ID','category_name',
            'attribute_value_pairs','date','policy_URL']
annotations_df.head()
```

```python
text=[]
for item in annotations_df.attribute_value_pairs:
    text.append(ast.literal_eval(item))
```

```python
text1=[]
for item in text:
    temptext=[]
    text1.append(temptext)
    for x,y in item.items():
        if y.get('selectedText')==None:
            temptext.append(str(y.get('selectedText')))
        else:
            temptext.append(y.get('selectedText'))
```

```python
text3=[]
for text in text1:
    text3.append(list(np.unique(text)))
```

```python
annotations_df['text']=text3
```

```python
annotations_df.head()
```

```python
for item in annotations_df.text:
    if 'None' in item:
        item.remove('None')
```

```python
annotations_df['plain_text']=annotations_df['text'].apply(lambda x: str(x).strip('[]'))
annotations_df.head()
```

```python
data=annotations_df[['category_name','plain_text']]
data
```

```python
data=data.loc[data['category_name']!= 'Other']
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
stop_words = set(stopwords.words('english')) 
```

```python
unique_class=[]
for i in data['category_name']:
    unique_class.append(i)
unique_class=set(unique_class)
```

```python
class_dict=dict(list (enumerate(unique_class)))
```

```python
class_numberdict = {}

for key, value in zip(class_dict.keys(), class_dict.values()):
    # Operations on key/value can also be performed.
    class_numberdict[value] = key
```

```python
class_numberdict
data['pretty_class']=data['category_name'].apply(lambda x: class_numberdict[x])
data.head()
```

```python
data.drop_duplicates(inplace=True)
```
