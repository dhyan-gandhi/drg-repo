# Deep Learning Tutorial with Keras
In this tutorial, we are going to use the Pima Indians onset of diabetes dataset. This is a standard machine learning dataset from the UCI Machine Learning repository. It describes patient medical record data for Pima Indians and whether they had an onset of diabetes within five years.

As such, it is a binary classification problem (onset of diabetes as 1 or not as 0). 

## Dataset Description
Number of Instances: 768   (500 Negative and rest is positive)

Number of Attributes: 8 

Each row of the data contains 8 features and a class
   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
   9. Class variable (0 or 1)

# Implement Deep Learning in 5 steps
## 1- Load Dataset
Ler me show you the first few rows

```python
!head pima-indians-diabetes.csv
```

```python
import numpy as np
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

## 2. Define The Model
i.e, you need to decide how many layers you want and how many neurons we will have in each layer.

We create a Sequential model and add layers one at a time until we are happy with our network topology. Sequential as it is layer by layer. For each of the layers, you will need to decide what kind of function will be inside the neurons of this layer. We have `tanh`, `sigmoid`, `RELU`.

**For the first layer only,** we need to explicitly tell our model how many inputs (features) we have. 8 in our case

We will create Five layers. 
1. First takes 8 inputs and consists of 20 neurons with RELU activation
2. Second and Third layer will have 15 neurons, with RELU
3. Fourth layer will have 9 neurons, with RELU
4. Fifth layer is the output layer which needs to decide 0 or 1, hence it will use sigmoid activation

```python
from keras.models import Sequential
from keras.layers import Dense
# create model
model = Sequential()
model.add(Dense(200, input_dim=8, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dense(90, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## 3.Compile The Model
In this step, we decide what optimization tool we will use and what computational hardware (CPU, GPU, .. etc)

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 4.Train the Model
We are now ready to feed the data into the model and let the model learn

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

## 5. Evaluate The Model
We will calculate the accuracy of our model using the scores variable

```python
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
